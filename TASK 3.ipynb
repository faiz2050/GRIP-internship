```python
#CDF Plot of petal length
counts,bin_edges=np.histogram(df[df['Species']=='Iris-setosa']['PetalLengthCm'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
```

```python
#compute CDF
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```

```python
from sklearn.cluster import KMeans
import sklearn.metrics as sm
from sklearn.model_selection import train_test_split
```

```python
df['Species'].replace({'Iris-virginica':0,'Iris-versicolor':1,'Iris-setosa':2},inplace=True)
```

```python
df.head()
```

```python
km=KMeans(n_clusters=3)
km
```

```python
X=(df[['PetalLengthCm','PetalWidthCm','SepalLengthCm','SepalWidthCm']])
y=(df[['Species']])
```

```python
X_test,X_train,y_test,y_train=train_test_split(X,y,test_size=0.25,random_state=0)
```
